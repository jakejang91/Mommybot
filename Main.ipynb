# Call modules for musicplayer, timemanger, and time


```python
import time
from t6modules.MusicPlayer import MusicPlayer
from t6modules.TimeManager import TimeManager
import serial

alarmused = False

# Common functions for all waiting loop. 
# If the time is reset time, the system will be reboot
# if the time is morning call time, play music
def commonFunctions(morningtime, resettime):
    global tm, alarmused
    print("The common Function Get : " + str([morningtime, resettime]))
    print("now time is : " + str(tm.l2t(time.time())))
    Ser = serial.Serial('/dev/ttyUSB0', 9600)
    # Ser.write(bytes(str(time.localtime(time.time()).tm_hour) + ' : ' + str(time.localtime(time.time()).tm_min), encoding='ascii'))
    
    try:
        try:
            if tm.l2t(time.time()) > resettime and tm.l2t(time.time()) < resettime +10:
                print("SYSTEM REBOOT")
                Ser.write(bytes('     Reboot', encoding='ascii'))
                alarmused = False
                return False
        except:
            print('reset error')

        try:    
            if tm.l2t(time.time()) > morningtime and tm.l2t(time.time()) < morningtime +5 and alarmused == False:
                print("MORNING CALL")
                mp = MusicPlayer('Wishful_Thinking.mp3')
                print("Set Music Player")
                Ser.write(bytes('     Wake Up!!', encoding='ascii'))
                mp.playMusic()
                alarmused = True
                time.sleep(30)
                mp.stopMusic()
                return True
        except:
            print('morning call error')

        return True
    except:
        print("common function problem")
        return True
```

```python
# Calculate your required sleep time from log file to achive your sleep goal.

def getRequiredSleepTime(mcall, sleep_goal, last_sleeps, weekday):
    print("GetRequired sleep time calculate")
    print(str(mcall) + " " + str(sleep_goal)+ " " + str(last_sleeps)+ " " + str(weekday) + "===")
    total_sleep_goal =  sleep_goal * 5 # unist : min, delta time
    today_purpose = (total_sleep_goal - last_sleeps*weekday)/(5-weekday)
    if today_purpose < sleep_goal:
        print("No need less sleep")
        today_purpose = sleep_goal
        
    print("goal is: " + str(mcall - today_purpose))
    return mcall - today_purpose #return timestamp
```

# MAIN: Initiate Mommybot

    - Below record shows how Mommybot records each time variables, and what Mommybot does using those variables.
    - Mommybot reset its time everyday at 4:00 pm.
    - Mommybot Mommybot uses time unit only minute. Therefore, at 4:00 pm = 960 m
        - it will continue until next day's 3:59 pm, which is 2399
    - As output, shows status of user, and timelogs, and status of Mommybot.
    - Used Variables
        1. T1-1: BT cconnection wating; when you arrived at home.
        2. T1-2: Wait until your go to bed
        3. T1-3: Wait until fall asleep
        4. T2-01: hours of Sleep time  yesterday
        5. T2-02: Average hours of sleep untill today
        6. T3: Weekday; Monday to Friday (0 ~ 4)
        

```python
print('------------ start Mommybot -----------------')

# functions to record logs(each time variable data)
from t6modules.LogManager import LogManager
while True:
    print("")
    weekday = time.localtime(time.time()).tm_wday
    nowhour = time.localtime(time.time()).tm_hour
    nowmin = time.localtime(time.time()).tm_min
    if weekday in [1, 2, 3, 4] :#and tm.isStart(time.time()): # only work in Tur - Fri, Start Condition
        try:
            # initializing
            isNotBreaked = True
            
            morningcall_time_today = morning_call_times[weekday]
            print("morningcalltime is " + str(morningcall_time_today)) 
            
            today_log = []
            try:
                logs = LogManager()
                print('Log manager loaded')
            except:
                print("problem of Log Manager")
                
            try:
                yd_sleep = logs.getYesterdaySleep()
                ls_sleep = logs.getLastdaysSleep()
                print("sleeps: " + str(yd_sleep) + " " + str(ls_sleep))
            except:
                print("No Yesterdays")
                
            try:
                #print("!!!!!!!!!!!!!!!!!")
                #print(morningcall_time_today)
                #print(sleep_goals)
                #print(ls_sleep)
                #print(weekday)
                today_required_sleep_start = getRequiredSleepTime(morningcall_time_today, sleep_goals, ls_sleep, weekday)
                print("today required sleep time start is : " + str(today_required_sleep_start))
            except:
                print("!")
            
            # T1-1 BT connection waiting
            while not isBtConnected(MAC_POCO):
                try:
                    isNotBreaked = commonFunctions(morningcall_time_today, timestamp_start)
                except:
                    print("error:t1-1 while loop")
                time.sleep(20)

                    
            if isNotBreaked:
                today_log.append(tm.l2t(time.time()))
                print("today log is :" + str(today_log))
                # Predict #1
                try:
                    pt2 = timepredictor.getPredictedTime(data = [today_log[0], yd_sleep, ls_sleep, weekday]
                                                         , variable = 't1-2')[0]
                    print('pt2 is: ' + str(pt2))
                    if pt2 > today_required_sleep_start-30:
                        print("You may sleep late :" +  str(pt2) + ">" + str(today_required_sleep_start-30))
                        pt2 = today_required_sleep_start-30
                        print("Go sleep :" +  str(pt2))
                    else:
                        print("OK :" +  str(pt2) + "<" + str(today_required_sleep_start-30))
                except:
                    print("Prediction 1 was failed")
                    isNotBreaked = False
            else:
                print("the system was breaked - time prediction")

            
            # T1-2 Wait until your go ti bed
            if isNotBreaked:
                status.reset()
                isSerialUsed = False
                while status.checkStatus() != 'not sleeping': #not sleeping
                    try:
                        print(status.checkStatus())
                        
                        try:
                            if tm.l2t(time.time()) > pt2 and isSerialUsed == False:
                                Ser.write(bytes('Go to bed now', encoding='ascii'))
                                mp2 = MusicPlayer('Time_to_bed.mp3')
                                mp2.playMusic()
                                #alarmused = TRUE
                                time.sleep(5)
                                mp2.stopMusic()

                                #serial.writeSerial('Go to bed now')
                                isSerialUsed = True
                        except:
                            print("serial error")
                        
                        try:
                            save_snapshot(imgpath)
                            status.updateLastStatue(sleepclassifier.getStatus(imgpath))
                        except:
                            print('status update error')

                        isNotBreaked = commonFunctions(morningcall_time_today, timestamp_start)

                        if isNotBreaked == False:
                            break

                        time.sleep(camera_time_interval)#camera_time_interval)
                    except:
                        print("error:T1-2 while loop")
                        isNotBreaked = False
                        break
                        

                # Now user goes to bed, record t1-2
                today_log.append(tm.l2t(time.time()))
                
                # Predict #2
                #Below line is to explain if user's when to sleep time is later than when he should sleep time, 
                #Mommybot will tell user to go to sleep at when he should sleep time.
                try:
                    pt3 = timepredictor.getPredictedTime(data = [today_log[0], today_log[1], yd_sleep, ls_sleep, weekday]
                                                         , variable = 't1-3')[0]
                    print('pt3 is: ' + str(pt3))
                    
                    if pt3 > today_required_sleep_start:
                        print("Warning :" +  str(pt3) + ">" + str(today_required_sleep_start))
                        pt3 = today_required_sleep_start
                    else:
                        print("OK :" +  str(pt3) + "<" + str(today_required_sleep_start))
                        
                except:
                    print("Prediction 2 was failed")
                    isNotBreaked = False
            else:
                print("the system was breaked - t1-2")

                
            time.sleep(60)
                

            # T1-3 : Wait until fall asleep
            if isNotBreaked:
                status.reset()
                isSerialUsed = False
                while status.checkStatus() != 'sleeping': #sleeping
                    try:
                        print(status.checkStatus())
                        
                        try:
                            if tm.l2t(time.time()) > pt3 and isSerialUsed == False:
                                mp3 = MusicPlayer('Time to sleep.mp3')
                                Ser.write(bytes('Stop Using Phone', encoding='ascii'))
                                mp3.playMusic()
                                #alarmused = TRUE
                                time.sleep(5)
                                mp3.stopMusic()
                                
                                #serial.writeSerial('Stop Using Phone')
                                isSerialUsed = True
                        except:
                            print("serial error") c  
                        
                        
                        try:
                            save_snapshot(imgpath)
                            status.updateLastStatue(sleepclassifier.getStatus(imgpath))
                        except:
                            print('status update error')

                        isNotBreaked = commonFunctions(morningcall_time_today, timestamp_start)


                        if isNotBreaked == False:
                            break

                        time.sleep(camera_time_interval)

                    except:
                        print("error:T1-3 while loop")
                        isNotBreaked = False
                        break

                # User is Sleeping! record t1-3
                today_log.append(tm.l2t(time.time()))
            else:
                print("the system was breaked - t1-3")

            ## save today timestamp
            if isNotBreaked:
                # t1-1,   2,      3, 
                # t-2-01, t-2-02  t3
                today_log.append(morningcall_time_today-today_log[2])
                today_log.append((morningcall_time_today-today_log[2]+ ls_sleep*weekday)/(weekday+1))
                today_log.append(weekday)
                print(today_log)
                
                logs.saveLog(today_log) ## list of 6 float
            else:
                print("the system is breaked - file save")
                
            while commonFunctions(morningcall_time_today, timestamp_start):
                time.sleep(10)
        except:
            print("Something Wrong on Tue - Fri, Restart!")
            time.sleep(60)
            
    elif weekday == 0 :#Monday, only save log,
        try:
            # initializing
            isNotBreaked = True
            
            morningcall_time_today = morning_call_times[weekday]
            print("morningcalltime is " + str(morningcall_time_today)) 
            
            today_log = []
            try:
                logs = LogManager()
                print('Log manager loaded')
            except:
                print("problem of Log Manager")
                
            try:
                yd_sleep = logs.getYesterdaySleep()
                ls_sleep = logs.getLastdaysSleep()
                print("sleeps: " + str(yd_sleep) + " " + str(ls_sleep))
            except:
                print("No Yesterdays")
                
            try:
                print("!!!!!!!!!!!!!!!!!")
                print(morningcall_time_today)
                print(sleep_goals)
                print(ls_sleep)
                print(weekday)
                today_required_sleep_start = getRequiredSleepTime(morningcall_time_today, sleep_goals, ls_sleep, weekday)
                print("today required sleep time start is : " + str(today_required_sleep_start))
            except:
                print("!")
            
            # T1-1 BT connection waiting
            while not isBtConnected(MAC_POCO):
                try:
                    isNotBreaked = commonFunctions(morningcall_time_today, timestamp_start)
                except:
                    print("error:t1-1 while loop")
                time.sleep(20)

                    
            if isNotBreaked:
                today_log.append(tm.l2t(time.time()))
                print("today log is :" + str(today_log))
                # Predict #1
                try:
                    pt2 = timepredictor.getPredictedTime(data = [today_log[0], yd_sleep, ls_sleep, weekday]
                                                         , variable = 't1-2')[0]
                    print('pt2 is: ' + str(pt2))
                    if pt2 > today_required_sleep_start-30:
                        print("You may sleep late :" +  str(pt2) + ">" + str(today_required_sleep_start-30))
                        pt2 = today_required_sleep_start-30
                        print("Go sleep :" +  str(pt2))
                    else:
                        print("OK :" +  str(pt2) + "<" + str(today_required_sleep_start-30))
                except:
                    print("Prediction 1 was failed")
                    isNotBreaked = False
            else:
                print("the system was breaked - time prediction")

            print(isNotBreaked)
            
    
            # T1-2 Wait until your go ti bed
            if isNotBreaked:
                status.reset()
                isSerialUsed = False
                while status.checkStatus() != 'not sleeping': #not sleeping
                    try:
                        try:
                            save_snapshot(imgpath)
                            status.updateLastStatue(sleepclassifier.getStatus(imgpath))
                        except:
                            print('status update error')

                        isNotBreaked = commonFunctions(morningcall_time_today, timestamp_start)

                        if isNotBreaked == False:
                            break

                        time.sleep(camera_time_interval)#camera_time_interval)
                    except:
                        print("error:T1-2 while loop")
                        isNotBreaked = False
                        break
                        

                # Now user goes to bed, record t1-2
                today_log.append(tm.l2t(time.time()))
                
                # Predict #2
                try:
                    pt3 = timepredictor.getPredictedTime(data = [today_log[0], today_log[1], yd_sleep, ls_sleep, weekday]
                                                         , variable = 't1-3')[0]
                    print('pt3 is: ' + str(pt3))
                    
                    if pt3 > today_required_sleep_start:
                        print("Warning :" +  str(pt3) + ">" + str(today_required_sleep_start))
                        pt3 = today_required_sleep_start
                    else:
                        print("OK :" +  str(pt3) + "<" + str(today_required_sleep_start))
                        
                except:
                    print("Prediction 2 was failed")
                    isNotBreaked = False
            else:
                print("the system was breaked - t1-2")

                
            time.sleep(60)
                

            # T1-3 : Wait until fall asleep
            if isNotBreaked:
                status.reset()
                isSerialUsed = False
                while status.checkStatus() != 'sleeping': #sleeping
                    try:
                        print(status.checkStatus())
                        try:
                            save_snapshot(imgpath)
                            status.updateLastStatue(sleepclassifier.getStatus(imgpath))
                        except:
                            print('status update error')
                        isNotBreaked = commonFunctions(morningcall_time_today, timestamp_start)
                        if isNotBreaked == False:
                            break
                        time.sleep(camera_time_interval)

                    except:
                        print("error:T1-3 while loop")
                        isNotBreaked = False
                        break

                # User is Sleeping! record t1-3
                today_log.append(tm.l2t(time.time()))
            else:
                print("the system was breaked - t1-3")

            ## save today timestamp
            if isNotBreaked:
                # t1-1,   2,      3, 
                # t-2-01, t-2-02  t3
                today_log.append(morningcall_time_today-today_log[2])
                today_log.append((morningcall_time_today-today_log[2]+ ls_sleep*weekday)/(weekday+1))
                today_log.append(weekday)
                print(today_log)
                
                logs.saveLog(today_log) ## list of 6 float
            else:
                print("the system is breaked - file save")
                
            while commonFunctions(morningcall_time_today, timestamp_start):
                time.sleep(10)
        except:
            print("Something Wrong on Monday, Restart!")
            time.sleep(60)

    else:
        print("Waiting until Tue or valid time")
        print("today is" + str(weekday))
        # waiting 1 hour
        time.sleep(3600)
```
